```python
#Juntando os dados
df_final = pd.DataFrame()
df_final = dados[['status','codigoQtdRegistroAbamdoc','codigoQtdRegistroAbbmItem']]

df_final = df_final.join( dadosseguradoVO_df[['anosFidelidadeCliente','codigoEstadoCivil','sexoSegurado', 'idade','tipoPessoaSegurado','resgatePontos','numeroSegmentoCliente']] )
df_final = df_final.join( dadosseguradoEndVO_df[['cidade','estado']] )

df_final = df_final.join( itensVeiculo_df[['codigoCategoriaTarifaria','quantidadePortasVeiculo','codigoCombustivelVeiculo','codigoUsoVeiculo']] )
df_final = df_final.join( itensVeiculoModelo_df[['codigoFamiliaVeiculo','especieVeiculo','codigoTipoVeiculo','codigoMarcaVeiculo']] )

try: 
    df_final = df_final.join( dadosSegmentoVO_df[['numeroSegmentoDocumento']] )
except:
    df_final['numeroSegmentoDocumento'] = 8
    pass

df_final = df_final.join( dadosApoliceVO_df[['codigoSucursal','numeroDigitoApolice','premioLiquidoPagoApolice','dataFinalVigencia']] )
df_final['dataFinalVigencia'] = pd.to_datetime(df_final['dataFinalVigencia'], unit='ms')

df_final = df_final.join( dadosCascoVO_df[['valorFranquia','valorPremioFinal','valorDiferencaPremioAnual', 'valorImportanciaSeguradaCasco', 'codigoClasseFranquia']] )

df_final['quantidadeUtilizacaoEstapar'] = quantidadeUtilizacaoEstapar_df.values

df_final['codigoClasseLocalizacao'] = codigoClasseLocalizacao_df.values

df_final['origemProposta'] = origem_df.values

df_final['codigoAgravamentoPremio'] = codigoAgravamentoPremio_df.values
df_final['quantidadeParcelas'] = quantidadeParcelas_df.values

df_final = df_final.join( dadosSinistroVO_df[['quantidadeSinistroAuto','quantidadeTotalSinistros']] )

df_final = df_final.join( dadosBonusVO_df[['codigoClasseBonus']] )

df_final = df_final.join( dadosFormaPagtoVO_df[['codigoFormaPagamento','cpfCorrentista']] )
```

```python
df_final['codigoEstadoCivil'] = df_final['codigoEstadoCivil'].astype(int)
df_final['codigoCategoriaTarifaria'] = df_final['codigoCategoriaTarifaria'].astype(int)
df_final['codigoUsoVeiculo'] = df_final['codigoUsoVeiculo'].astype(int)
df_final['codigoFamiliaVeiculo'] = df_final['codigoFamiliaVeiculo'].astype(int)

df_final['especieVeiculo'] = df_final['especieVeiculo'].astype(int)
df_final['codigoTipoVeiculo'] = df_final['codigoTipoVeiculo'].astype(int)
df_final['codigoMarcaVeiculo'] = df_final['codigoMarcaVeiculo'].astype(int)
df_final['numeroSegmentoDocumento'] = df_final['numeroSegmentoDocumento'].astype(int)

df_final['quantidadePortasVeiculo'] = df_final['quantidadePortasVeiculo'].astype(int)
df_final['codigoCombustivelVeiculo'] = df_final['codigoCombustivelVeiculo'].astype(int)

df_final['numeroSegmentoCliente'] = df_final['numeroSegmentoCliente'].astype(int)
df_final['resgatePontos'] = df_final['resgatePontos'].astype(int)

df_final['codigoClasseBonus'] = df_final['codigoClasseBonus'].astype(int)
```

```python
df_final.dtypes
```

```python
df_final['IDHM'] = df_final['IDHM'].map(lambda x: x.replace(",","."))
df_final['IDHM_E'] = df_final['IDHM_E'].map(lambda x: x.replace(",","."))
df_final['IDHM_L'] = df_final['IDHM_L'].map(lambda x: x.replace(",","."))
df_final['IDHM_R'] = df_final['IDHM_R'].map(lambda x: x.replace(",","."))
df_final['I_ESCOLARIDADE'] = df_final['I_ESCOLARIDADE'].map(lambda x: x.replace(",","."))

df_final['correntista'] = [1 if x != None else  0 for x in df_final['cpfCorrentista'] ]
df_final['resgatePontos'] = [1 if x > 0 else  0 for x in df_final['resgatePontos'] ]
```

```python
df_final['IDHM'] = df_final['IDHM'].astype(float)
df_final['IDHM_E'] = df_final['IDHM_E'].astype(float)
df_final['IDHM_L'] = df_final['IDHM_L'].astype(float)
df_final['IDHM_R'] = df_final['IDHM_R'].astype(float)
df_final['I_ESCOLARIDADE'] = df_final['I_ESCOLARIDADE'].astype(float)
```

```python
#Faixa de Idade
df_final['faixaIdade'] = [1 if ( x >=18 and x <=24 ) else 2 if ( x >=25 and x <=60 ) else 3 for x in df_final['idade'] ]
del df_final['idade']

#Sexo
df_final['sexoSegurado'] = [1 if x == 'F' else  0 for x in df_final['sexoSegurado'] ]
df_final['sexoSegurado'] = df_final['sexoSegurado'].astype(int)

#TipoPessoa
df_final['tipoPessoaSegurado'] = [1 if x == 'F' else  0 for x in df_final['tipoPessoaSegurado'] ]
df_final['tipoPessoaSegurado'] = df_final['tipoPessoaSegurado'].astype(int)

#Faixa de Preços
menor =  df_final['premioLiquidoPagoApolice'].describe()[4]
media =  df_final['premioLiquidoPagoApolice'].mean()
df_final['faixaPrecoLiquido'] = [1 if ( x >=0 and x <= menor) else 2 if ( x > menor and x <=media ) else 3 for x in df_final['premioLiquidoPagoApolice'] ]

#Faixa de Casco
menor =  df_final['valorFranquia'].describe()[4]
media =  df_final['valorFranquia'].mean()
df_final['faixaPrecoFranquia'] = [1 if ( x >=0 and x <= menor) else 2 if ( x > menor and x <=media ) else 3 for x in df_final['valorFranquia'] ]

menor =  df_final['valorPremioFinal'].describe()[4]
media =  df_final['valorPremioFinal'].mean()
df_final['faixaPrecoPremioFinal'] = [1 if ( x >=0 and x <= menor) else 2 if ( x > menor and x <=media ) else 3 for x in df_final['valorPremioFinal'] ]

```

```python
#Diferenca de Premios
df_final['diferencaPremioAntNovo'] = df_final.valorPremioPagoAtual - df_final.premioLiquidoPagoApolice
menor =  df_final['diferencaPremioAntNovo'].describe()[4]
media =  df_final['diferencaPremioAntNovo'].mean()
df_final['faixaDiferencaValores'] = [1 if ( x <= menor) else 2 if ( x > menor and x <=media ) else 3 for x in df_final['diferencaPremioAntNovo'] ]
```

```python
#Montando a variavel target
df_final['calc1'] = df_final['status'] + df_final['codigoQtdRegistroAbamdoc'] + df_final['codigoQtdRegistroAbbmItem']
df_final['renovou'] = [1 if (x == '1911' or x == '1901' or x == '1910') else 0 for x in df_final['calc1'] ]
df_final['gerouSinistro'] = [1 if x > 0 else 0 for x in df_final['quantidadeSinistroAuto'] ]
```

```python
del df_final['calc1']
del df_final['status']
del df_final['codigoQtdRegistroAbamdoc']
del df_final['codigoQtdRegistroAbbmItem']
del df_final['dataFinalVigencia']
del df_final['cpfCorrentista']
del df_final['quantidadeSinistroAuto']
del df_final['quantidadeTotalSinistros']       
```
